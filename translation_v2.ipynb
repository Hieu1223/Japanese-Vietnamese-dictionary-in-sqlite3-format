```python
class Token:
    def __init__(self,word,reading,type,desc):
        self.word = word
        self.reading = reading
        self.type = type
        self.desc = desc
    def __str__(self):
        print(f"word {self.word} reading {self.reading} type {self.type} desc {self.desc}")
```

```python
first_pass_exception = []
first_pass_result = []
for line in sentences[kanjiStart:-1]:
    try:
        key, desc = line.split("\t-")
        first_pass_result.append([key,desc])
    except:
        first_pass_exception.append(line)
```

```python
first_pass_exception
```

```python
second_pass_exception = []
for line in first_pass_exception:
    try:
        if "\t@" in line:
            word, desc = line.split("\t@",1)
            reading, meaning = line.split(" - ",1)
            results.append(Token(word=word, reading=reading,type="phrase",desc=meaning))
            continue
        if "\t*" in line:
            word, desc = line.split("\t*",1)
            wtype,meaning = line.split("-",1)
            results.append(Token(word=word,reading="*", type=wtype,desc=meaning))
            continue
        second_pass_exception.append(line)
    except:
        second_pass_exception.append(line)
```
